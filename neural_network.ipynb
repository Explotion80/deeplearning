## Implementacja sieci neuronowej

```
import numpy as np

np.random.seed(0)

X = np.array([[1.0, 0.7]])
y_true = np.array([1.80])

```

```
def init_parameters(n_x, n_h, n_y):
    W1 = np.random.randn(n_h, n_x)
    W2 = np.random.randn(n_h, n_y)
    return W1, W2

def forward_prop(X, W1, W2):
    # propagacja w prz√≥d
    H = np.dot(X, W1)
    y_pred = np.dot(H, W2)
    return H, y_pred

def calculate_error(y_true, y_pred):
    return y_pred - y_true

    # propagacja wsteczna
def backprop(X, W1, W2, learning_rate=0.01, iters=1000, precision=0.0000001):

    H, y_pred = forward_prop(X, W1, W2)

    for i in range(iters):
        error = calculate_error(y_true, y_pred)
        W2 = W2 - learning_rate * error * H.T
        W1 = W1 - learning_rate * error * X.T * W2.T

        _, y_pred = forward_prop(X, W1, W2)

        print('Iter {}, y_pred: {}, error: {}' .format(i, y_pred[0][0], calculate_error(y_true, y_pred[0][0])))

        if abs(error) < precision:
         break

    return W1, W2 

def predict(X, W1, W2):
  _, y_pred = forward_prop(X, W1, W2)
  return y_pred

def build_model():

  # inicjalizacja wag
  W1, W2 = init_parameters(2, 2, 1)

  # propagacja wsteczna
  W1, W2 = backprop(X, W1, W2)

  model = {'W1': W1, 'W2': W2}

  return model
    
```
